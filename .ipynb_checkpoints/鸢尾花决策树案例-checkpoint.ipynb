# 读取数据

```python
import pandas as pd
```

```python
df = pd.read_csv(r'data/iris.csv')
df.head()
```

# 决策树分类

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split # 取一部分数据作为测试数据集
from sklearn.metrics import accuracy_score # 模型评估
```

```python
# 划分数据
data = df[df.columns[:4]]
labels = df[df.columns[-1]]
xtrain,xtest,ytrain,ytest = train_test_split(data,labels,test_size=0.25,stratify=labels,random_state=14)
```

```python
# 决策树模型可视化并保存图片
from sklearn import tree
import pydotplus
from PIL import Image
def save_DT(model,fn):
    dot_data = tree.export_graphviz(model, out_file=None,
                                    feature_names=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid'],
                                    class_names=['0', '1','2'],
                                    filled=True, rounded=True,
                                    special_characters=True)
    graph = pydotplus.graph_from_dot_data(dot_data)
    with open(fn, 'wb') as f:
        f.write(graph.create_png())
    return Image.open(fn)
```

```python
model_1 = DecisionTreeClassifier()
model_1.fit(xtrain,ytrain)
pred = model_1.predict(xtest)
accuracy_score(ytest,pred)
```

```python
save_DT(model_1,r'DTree.png')
```

从上图看出，默认参数情况下模型训练集数据最后的分类的gini指数都为0。这说明模型对训练数据的分类太干净了，很可能出现过拟合的现象，所以需要对决策树进行剪枝操作。     


**sklearn决策树中重要的参数：**
>`max_depth`树的最大深度（分割点的个数），常用于减少模型复杂度防止过拟合       
>`min_samples_leaf`每个叶子的最大个数           
>`max_leaf_nodes`树中叶子的最大个数               
 
**一般只需要调整max_depth就可以防止决策树过拟合**

```python
model_2 = DecisionTreeClassifier(max_depth=4) # 限制树的深度
model_2.fit(xtrain,ytrain)
pred = model_2.predict(xtest)
accuracy_score(ytest,pred)
```

```python
save_DT(model_2,r'DTree(1).png')
```
